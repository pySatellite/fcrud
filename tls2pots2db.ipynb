```python
def get_tle(page: int):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    r = requests.get(f"https://data.ivanstanojevic.me/api/tle/?page={page}", headers=headers)
    data = r.json()
    return data.get('member')
    
```

```python
def call_post(mem):
    import requests

    url = "http://localhost:8000/orbits"
    headers = {
        'accept': 'application/json',
        'Content-Type': 'application/json'
    }
    
    data = {
        "name": mem.get('name'),
        "date": mem.get('date'),
        "tle_id": int(mem.get('satelliteId')),
        "line1": mem.get('line1'),
        "line2": mem.get('line2')
    }
    
    # Make a POST request
    response = requests.post(url, headers=headers, json=data)    

```

```python
for i in range(2, 101):
    print(str(i).center(10, "#"))
    members = get_tle(page=i)
    for o in members:
        call_post(o)
        print(o.get('name'))
```
